## Build decision tree model to predict survival based on certain parameters
Using following columns build a model to predict if person would survive or not,
.

###### Pclass,Sex,Age,Fare.
dataset - https://github.com/codebasics/py/blob/master/ML/9_decision_tree/Exercise/titanic.csv 

```python
import pandas as pd 
df = pd.read_csv("titanic.csv")
df
```

```python
df.head(10)
```

```python
df = df.drop(['Ticket','Cabin','Embarked'],axis =1)
df
```

```python
df = df.drop(['SibSp','Parch'],axis =1)
df
```

```python
inputs = df.drop('Survived',axis= 'columns')
```

```python
target = df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass = LabelEncoder()
le_Sex = LabelEncoder()
le_Age =LabelEncoder()
le_Fare =LabelEncoder()
    
```

```python
inputs['Pclass_n'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['Fare_n'] = le_Fare.fit_transform(inputs['Fare'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Name','Pclass','Sex','Age','Fare'],axis = 'columns')
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(inputs_n,target)
```

```python
# Score of Model - it is 0.97 means 97%
model.score(inputs_n,target)
```

#### giving detils of passenger if array gives '1' as output passenger is survived & if gives '0' not survived !
##### Class , Gender , Age , Fare 

```python
model.predict([[0,0,51,207]]) #passenger survived 
```

```python
model.predict([[0,1,0,180]]) # passenger survived
```

```python
model.predict([[2,1,42,30]])  # passenger not survived
```

```python
model.predict([[2,1,1,500]]) # passenger not survived
```
